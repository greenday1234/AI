```python
import os
import subprocess

def convert_audio_ffmpeg(input_dir, output_dir, output_format):
    # 입력 디렉토리에서 오디오 파일들을 찾아 변환합니다
    for filename in os.listdir(input_dir):
        if filename.endswith((".wav", ".mp3", ".flac", ".ogg")):  # 변환할 파일 확장자들을 지정
            input_path = os.path.join(input_dir, filename)
            output_filename = os.path.splitext(filename)[0] + f".{output_format}"
            output_path = os.path.join(output_dir, output_filename)

            # ffmpeg 명령어 호출
            command = f"ffmpeg -i \"{input_path}\" \"{output_path}\""
            subprocess.run(command, shell=True, check=True)
            print(f"Converted {filename} to {output_filename}")
```

```python
input_directory = "/Users/ichanhui/Desktop/archive"
output_directory = "/Users/ichanhui/Desktop/test"
desired_format = "wav"

if not os.path.exists(output_directory):
    os.makedirs(output_directory)
    
convert_audio_ffmpeg(input_directory, output_directory, desired_format)
```
